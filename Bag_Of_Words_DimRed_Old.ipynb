```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
np.random.seed(2)
```

```python
def normalize_df(_df):
    return _df / _df.values.max(axis=1).reshape(-1, 1)

df = pd.read_csv("bag_of_words.csv").drop('Unnamed: 0', axis=1)
df = normalize_df(df)
df.head()
```

```python
model_pca = PCA(n_components=400)
data_pca = model_pca.fit_transform(df)
```

```python
sum(model_pca.explained_variance_ratio_)
```

```python
plt.plot(model_pca.explained_variance_ratio_)

plt.xlabel('Principal component index')
plt.ylabel('Explained variance ratio')
plt.show()
```

```python
for perp in [5, 11, 21, 31, 41]:
    model_tsne = TSNE(random_state=0, verbose=1, perplexity=perp)
    data_tsne = model_tsne.fit_transform(data_pca)

    plt.figure(figsize=(10, 7))
    plt.scatter(data_tsne[:,0], data_tsne[:,1])
    plt.title(f'Low Dimensional Representation of Vacancies. Perplexity = {perp}')
    
plt.show()
```
